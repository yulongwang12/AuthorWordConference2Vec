```python
import numpy as np
from gensim.models import Doc2Vec
import time
from collections import defaultdict
from gensim.models.doc2vec import LabeledSentence
from random import shuffle
```

```python
papers = np.load('extr-titles-abstract.npy')
```

```python
net_info = defaultdict(lambda: ([],[]))

for paper in iter(papers):
    # paper[i] = [interest list, authors, venue]
    interests = paper[0]
    authors = paper[1]
    venue = paper[2]
    for author in iter(authors):
        for interest in iter(interests):
            net_info[author][0].append(interest)
        
        net_info[author][1].append(venue)
```

```python
len(net_info.keys())
```

```python
doc_model = Doc2Vec(dm=1, dm_concat=1, size=100, \
                window=5, negative=5, hs=0, min_count=2, workers=-1)
```

```python
author_docs = []
author_venues = []
counter = 0
for author, doc_venue in net_info.iteritems():
    author_docs.append(LabeledSentence(doc_venue[0], [author+'_a']))
    author_venues.append(LabeledSentence(doc_venue[1], [author+'_a']))
```

```python
doc_model.build_vocab(author_docs)
```

```python
st = time.time()
for i in range(100):
    doc_model.train(shuffle(author_docs))
    print "iter %d, elapsed for %f" % (i, time.time() - st)
    st = time.time()
```

```python
author_conf_model = Doc2Vec(dm=1, dm_concat=1, size=100, \
                window=5, negative=5, hs=0, min_count=2, workers=-1)
```

```python
author_conf_model.build_vocab(author_venues)
```

```python
st = time.time()
for i in range(100):
    author_conf_model.train(shuffle(author_venues))
    print "iter %d, elapsed for %f" % (i, time.time() - st)
    st = time.time()
```
