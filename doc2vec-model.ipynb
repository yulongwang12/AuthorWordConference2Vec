```python
import numpy as np
from gensim.models import Doc2Vec
import time
from collections import defaultdict
from gensim.models.doc2vec import LabeledSentence
from random import shuffle
```

```python
papers = np.load('extr-titles-abstract.npy')
```

```python
net_info = defaultdict(lambda: [])

for paper in iter(papers):
    # paper[i] = [interest list, authors, venue]
    interests = paper[0]
    authors = paper[1]

    for author in iter(authors):
        for interest in iter(interests):
            net_info[author].append(interest)
```

```python
len(net_info.keys())
```

```python
doc_model = Doc2Vec(dm=1, dm_concat=1, size=100, \
                window=5, negative=5, hs=0, min_count=2, workers=-1)
```

```python
author_docs = []
counter = 0
for author, doc_venue in net_info.iteritems():
    author_docs.append(LabeledSentence(doc_venue, [author+'_a']))
```

```python
doc_model.build_vocab(author_docs)
```

```python
st = time.time()
for i in range(1000):
    doc_model.train(shuffle(author_docs))
    print "iter %d, elapsed for %f" % (i, time.time() - st)
    st = time.time()
```
